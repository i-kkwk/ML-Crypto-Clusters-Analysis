```python
#Dependencies
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from time import time
import matplotlib.lines as lines
%matplotlib inline

import warnings
warnings.filterwarnings('ignore') 
```

```python
#Load dataset from csv
file_path = "Resources/crypto_data.csv"
crypto_data = pd.read_csv(file_path, index_col = 0)
crypto_data.head()
```

```python
#Create copy file
crypto_df = crypto_data.copy()
```

```python
#Look up the shape of dataset
crypto_df.shape
```

```python
(crypto_df["IsTrading"] == True)
```

```python
#filter for currencies that are currently being traded
data = (crypto_df["IsTrading"] == True)
crypto = crypto_df[data]
```

```python
#Check value again to see only "True" there.
crypto["IsTrading"].value_counts()
```

```python
crypto_df1 = crypto[crypto.Algorithm.isna() == False]
crypto_df1.shape
```

```python
#Drop any null values in dataset
crypto_df1.dropna(inplace = True)
crypto_df1
```

```python
#Drop "IsTrading" column
crypto_df = crypto_df1.drop("IsTrading", axis=1)
crypto_df
```

```python
#Filter for cryptocurrencies that have been mined.
#Set the total coins mined greater than 0
crypto_df = crypto_df[(crypto_df.TotalCoinsMined > 0)]
crypto_df
```

```python
#Create individual DataFrame for "CoinName" seperately that could be use later
crypto_name = pd.DataFrame(crypto_df.CoinName)
```

```python
#Drop "CoinName" from original dataset
crypto_df = crypto_df.drop("CoinName", axis = 1)
crypto_df
```

```python
#Use get.dummies to convert remaing features with test values into numerical data
X = pd.get_dummies(crypto_df, columns = ["Algorithm", "ProofType"])
print (X.shape)
X
```

```python
#Standardize dataset, fit_transform
scaler = StandardScaler()
crypto_scaled = scaler.fit_transform(X)
crypto_scaled[0:1]
```

```python
crypto_scaled.shape
```

```python
#Perform dimensionality reduction with PCA
pca = PCA(0.90)
crypto_pca = pca.fit_transform(crypto_scaled)
crypto_pca.shape
print ( "Components = ", pca.n_components_ , ";\nTotal explained variance = ",
      round(pca.explained_variance_ratio_.sum(),3)  )

```

```python
# Using the output of PCA as input for t-SNE
%time
tsne_feature = TSNE(random_state =13, verbose=0, perplexity=40, n_iter=1000).fit_transform(crypto_pca)
```

```python
#Create a scatter plot of the t-SNE output
pca_tsne = pd.DataFrame()
pca_tsne['x'] = tsne_feature[:,0]
pca_tsne['y'] = tsne_feature[:,1]
sns.scatterplot(data = pca_tsne, x = "x", y = "y", alpha=0.6)
```

```python
#Create an elbow plot to identify the best number of clusters.
#Use for-loop to determine the inertia for each k between 1 to 10

kstore = []

k = list(range(1,10,2))

for i in k:
    kmean = KMeans(n_clusters = i, random_state =40)
    kmean.fit(crypto_pca)
    kstore.append(kmean.inertia_)

eb_data = {"k": k, "kstore": kstore}
eb_data = pd.DataFrame(eb_data)
```

```python
plt.figure(figsize = (12, 8))
plt.title("The Elbow Method", size = 25, y = 1.0, fontname = "monospace")
plt.grid(color = "gray", linestyle = ':', axis = "y", alpha = 0.8, zorder = 0,  dashes = (1,7))

a = sns.lineplot(x = k, y = kstore, color = '#336b87', linewidth = 3)
sns.scatterplot(x = k, y = kstore, color = '#336b87', s = 60, edgecolor = 'black', zorder = 5)

plt.ylabel('Inertia', size = 14, fontname = 'monospace')
plt.xlabel('Number of clusters', size = 14, fontname = 'monospace')
plt.xticks(size = 12, fontname = 'monospace')
plt.yticks(size = 12, fontname = 'monospace')

plt.show()
```
